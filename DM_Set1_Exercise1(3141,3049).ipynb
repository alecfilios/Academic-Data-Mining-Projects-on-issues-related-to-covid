#### 1A. Reservoir Sampling for K size capacity. 
Given the description of the Reservoir Sampling problem we can subtract the information needed to solve it. First of all we have a capacity of size K for our samples to be stored.Which means that in any given time the used memory would be maximum of O(K). The total number of elements in the stream is unknown ,till the end of it which then will be called N. The algorithm cannot look back at previous items. Therefore we can proceed to the analysis of the algorithm

##### 1.
The algorithm consists of an input stream, of unknown size N, and we want to produce an output array of size K that contains distinct random elements of the input stream, without requiring any additional memory for elements selected by the sample.  At first we need to fill the empty reservoir with the first K elements of the input stream. Then each of the following elements has a probability to replace one of the filled positions in the reservoir.We pick a random index from 0 to i and if the picked index is one of the first K indexes, we replace the element at picked index of the output array.When the input stream is finished or the procedure is terminated, each and every element had the same chance and probability to be included in the final reservoir.The result of the algorithm is the reservoir itself. 
Note that the samples are NOT replaced in the input stream if they get discarded.

##### 2.
As we claimed in the previous answer , every element from all N elements we encountered ,had the same probability K/N to be part of the final result.Now we are going to analyse the probabilities behind that claim and prove it.
The first K elements have of course a probability of 1 to be picked.
After them the rest N-K samples have a probability K/i  , where i  is the index of the item currently under consideration.As it was already mentioned the procedure consists of a random number between 1 and i. And if the number is less or equal to K then the sample is picked. Thus the probability K/i results.
Similarly, at each iteration every element of the reservoir array is chosen to be replaced with probability 1/K x K/i = 1/i .Which K/i is the probability  of the sample currently under consideration and 1/K the probability of a candidate position in the array to be replaced both occur.
To prove that this solution works perfectly, we must prove that the probability that any item constitutes in the final array is K/N.With the above information we can proceed to the conclusion of our proof, but first  let us divide it in two cases as first k items are treated differently.

**Case1: For each item with index i , where** $ K+1 \leq i \leq N $ 

The probability that the i-th item is in the final reservoir = The probability that the item is picked (K/i ) and replaced an existing one and none of the following items replaced it, both occur : 

$$ (\frac{K}{i})(1-\frac{1}{i +1})(1-\frac{1}{i + 2})…(1-\frac{1}{N}) =(\frac{K}{i})(\frac{i}{i + 1})(\frac{i + 1}{i + 2})…(\frac{N-1}{N}) = \frac{K}{N} $$

**Case2: For each item with index i , where** $ 1 \leq i \leq K $

As for the first K items , the probability of them to stay on the final array till the end = probability of 1 and the probability of not being replaced by all the K+1 to N items that followed both occur, which is: 

$$ \prod_{i=K+1}^N(1-\frac{1}{i})= \prod_{i=K+1}^N(\frac{i-1}{i})=(\frac{Κ}{Κ + 1})(\frac{Κ + 1}{Κ + 2})...(\frac{N-1}{N})=\frac{K}{N} $$

In conclusion each and every of the N items of the stream input stream has the same probability K/N to be in the final array.And that’s how Reservoir Sampling algorithms works.

#### 1B. Weighted Reservoir Sampling for K size capacity . 
The problem of random sampling without replacement calls for the selection of a random sample out of a population of size N.As we saw in 1A if all items have the same probability to be selected, the problem is known as  RS. In weighted random sampling (WRS) the items are weighted and the probability of each item to be selected is determined by its relative weight $ {w_i} $ and the sum of all the N weights is called W and it's unknown till the end of the stream. WRS can be defined with the following algorithm base on the Reservoir Sampling we mentioned before.

1. In the following algorithm each element $ x_i $ with $ 1 \leq i \leq N $ that comes from the stream has it's own weight $ w_i $ which is a portion of the overall weight $ W=\sum\limits_{i=1}^N w_i $ .We now fill the reservoir array with the first element.At this point we've set up the core of the process and it's time to proceed to the procedure that makes the decision about which element is suited to replace an existing one on the reservoir array. In every iteration we add the weight of the current item we got from the stream to the current sum of the weights $ W_i=\sum\limits_{j=1}^i w_j $. [Note: not the same as overall W] . By a probability dependent on the weight of the item currently under consideration , the item replaces the one that is in the reservoir. The items have a weighted probability to be in the final reservoir. These are the steps of the Weightened Reservoir Algorithm.

2. In the description of the Algorithm we claimed that the probability for every item in the input stream to be in the final reservoir is dependent to it's own weight. More specifically that probability is $ \frac{w_i}{W} $ where W is the overall weight for all the N elements. To make this assumption we are going to analyse some of the probabilities that appear in the problem.
First of all the probability of the first item to be in the reservoir is of course 1. For the rest of the items the probability to be picked and to replace the current element is: $$ \frac{w_i }{W_i} $$ where: $ W_i=\sum\limits_{i=1}^i w_j $ .Finally the probability of an item to be in the final reservoir = probability of the item to be picked in the first place and the probability of **none** other item after it to replace it , both to occur:

$$ \frac{w_i}{W_{i}}\times (1 - \frac{w_{i+1}}{W_{i+1}})\times (1 - \frac{w_{i+2}}{W_{i+2}})...\times(1 - \frac{w_{N}}{W}) = $$ 
$$ $$
$$ \frac{w_i}{\sum_{j=1}^i w_j} \times (1 - \frac{w_{i+1}}{\sum_{j=1}^{i+1} w_j})\times (1 - \frac{w_{i+2}}{\sum_{j=1}^{i+2} w_j})...\times(1 - \frac{w_{N}}{\sum_{j=1}^{n}w_j}) = $$ 
$$ $$
$$ \frac{w_i}{\sum_{j=1}^i w_j} \times \frac{\sum_{j=1}^{i+1} w_j - w_{i+1}}{\sum_{j=1}^{i+1} w_j} \times \frac{\sum_{j=1}^{i+2} w_j - w_{i+2}}{\sum_{j=1}^{i+2} w_j}... \times \frac{W - w_n}{W} = $$ 
$$ $$
$$ \frac{w_i}{\sum_{j=1}^i w_j} \times \frac{\sum_{j=1}^i w_j}{\sum_{j=1}^{i+1} w_j} \times \frac{\sum_{j=1}^{i+1} w_j}{\sum_{j=1}^{i+2} w_j}... \times \frac{\sum_{j=1}^{N-1} w_j}{W} =  \frac {w_i}{W}$$
$$ $$
In conclusion each and every of the N items of the stream input stream has the same probability $ \frac {w_i}{W} $ to be in the final array.And that’s how Weighted Reservoir Sampling algorithms works.
